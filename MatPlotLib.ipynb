# **Matplotlib**
A powerful tool capable of producing complex publication-quality figures with fine layout control in two and three dimensions . It is an older library , so many libraries are built on top of it and use its syntax


```python
#The inline flag will use the appropriate backend to make figures apprear inline in the notebook 
%matplotlib inline 

import pandas as pd
import numpy as np

# `plt` is an alias for `matplotlib.pyplot` module
import matplotlib.pyplot as plt

# Import seaborn library (wrapper of matplotlib)
import seaborn as sns
```

```python
#Load car loan data into pandas dataframe from a csv file 
df=pd.read_csv('table_i.csv')

#View first 5 rows of dataframe 
df.head()
```

```python
#Check that we do not have NaN in our dataframe 
df.info()
#Plotting the data that contains NaN is difficult
```

```python
#We are going to graph month number on X axis , interest paid & principle paid on Y axis

#Turning pandas attributes to numpy arrays and assigning them to variables on left of '=' 
month_number=df.loc[:,'month'].values
interest_paid=df.loc[:,'interest_paid'].values
principal_paid=df.loc[:,'principal_paid'].values
```

```python
#Chnaging / adding marker
plt.style.use('ggplot')

plt.plot(month_number,interest_paid,marker='.',markersize=10)
plt.plot(month_number,principal_paid,marker='.',markersize=10)
```

# **Changing colors**
The c parameter accepts the string 



```python
plt.style.use('ggplot')

plt.plot(month_number,interest_paid,c='k',marker='.',markersize=10)
#k is for black color 
plt.plot(month_number,principal_paid,c='b',marker='.',markersize=10)
#b is for blue

#Hex strings can also be used , RGB tuples too can be used 
```

# **Matlab Style vs Object Oriented Style**
Matplotlib has two different types of syntax.

1) Matlab style - This is scripted interface designed to feel like MATLAB .Matplotlib maintains a pointer to current figure and sends commands to it

2) Object Oriented - This is used in situations where you want more control over figure.

```python
#MATLAB STYLE 
plt.style.use('ggplot')

plt.plot(month_number,interest_paid,marker='.',markersize=10)
plt.plot(month_number,principal_paid,marker='.',markersize=10)
```

```python
#OBJECT ORIENTED 

#Tuple unpacking 
x,y=(3,9)

fig,axes=plt.subplots(nrows=1,ncols=1)
axes.plot(month_number,interest_paid,c='k')
axes.plot(month_number,principal_paid,c='b')
```

# **Setting plot titles , labels & limits**


```python
#Setting x and y limits 

plt.plot(month_number,interest_paid,marker='.',markersize=10)
plt.plot(month_number,principal_paid,marker='.',markersize=10)
plt.xlim(left=1,right=65)
plt.ylim(bottom=0,top=800)

#This is not most practical way of changing limits
```

```python
plt.style.use("seaborn")
plt.plot(month_number,interest_paid,marker='.',markersize=10)
plt.plot(month_number,principal_paid,marker='.',markersize=10)
#Setting x and y labels & changing fontsize
plt.xlabel('Month',fontsize=22)
plt.ylabel('Dollars',fontsize=22)
#Setting title
plt.title("Interest and Principal Paid each month",fontsize=22)

#Changing tick font size
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```

```python
#Changing labels , fonts etc in Object oriented type 
fig,axes=plt.subplots(nrows=1,ncols=1)
axes.plot(month_number,interest_paid,c='k')
axes.plot(month_number,principal_paid,c='b')
#Setting x and y limits 
axes.set_xlim(left=1,right=70);
axes.set_ylim(bottom=1,top=900);
#Setting x and y labels 
axes.set_xlabel('Month',fontsize=20)
axes.set_ylabel('Dollars',fontsize=20)
#Setting the title 
axes.set_title("Interest and Principal Paid each month",fontsize=22)

#Changing tick font size
axes.tick_params(axis='x',labelsize=20)
axes.tick_params(axis='y',labelsize=20)
```

# Grids

```python
plt.style.use("classic")
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
plt.grid(c='g')#puts grid lines
```

```python
#Only horizontal grid lines

plt.style.use("classic")
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
plt.grid(axis='y')
```

```python
#Only vertical grid lines

plt.style.use("classic")
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
plt.grid(axis='x',c='g')
```

# **Plot Legends**
They assist in assigning meanings to various plot elements 

```python
plt.style.use("seaborn")
plt.plot(month_number,interest_paid,c='k',label='Interest')
plt.plot(month_number,principal_paid,c='b',label='Principal')
plt.legend(loc='center right')

plt.savefig('new.png',dpi=300)
```

# **Boxplot**
A standardized way of displaying distribution of data based on five number summary 


```python
from google.colab import files
uploaded=files.upload()
```

```python
cancer_df=pd.read_csv('wisconsinBreastCancer.csv')
cancer_df.head()
```

```python
#Plotting using pure Matplotlib

malignant=cancer_df.loc[cancer_df['diagnosis']=='M','area_mean'].values
benign=cancer_df.loc[cancer_df['diagnosis']=='B','area_mean'].values

plt.boxplot([malignant,benign],labels=['M','B'])
```

```python
#Plotting using pandas
cancer_df.boxplot(column='area_mean',by='diagnosis')
```

```python
#Plotting using seaborn
sns.boxplot(x='diagnosis',y='area_mean',data=cancer_df)
```
